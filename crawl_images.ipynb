````python
def rename(group_name, folder): 
    # Rename all files in folder with group_name
    i = 0
    group_name = group_name + '_'
    
    for f in os.listdir(folder):
        old_name, file_extension = os.path.splitext(f)
        new_name = group_name + str(i) + file_extension
          
        os.rename(folder + '\\' + f, folder + '\\' + new_name)
        i += 1
````

````python
# Rename files
folder = 'C:\Data\jack-russell-terrier'
rename('jack_russel_terrier', folder)
````
